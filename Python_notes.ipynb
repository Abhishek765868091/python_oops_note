# MULTI-THREADING

Thread :- A thread is smallest unit of execution inside a program.By default every python program runs on a single thread.

Multithreading :- running multiple threads at same time inside one program.

without multithreading tasks run one after another (sequentially).

_with multithreading task run side by side -- saving time

```python
import threading
import time
lock=threading.Lock
def task1():
    with lock():
        print("task1 is start")
        time.sleep(1)
    with lock():
        print("task1 is end")
def task2():
    with lock():
        print("task2 is start")
        time.sleep(1)
    with lock():
        print("task2 is end")
t1=threading.Thread(target=task1)
t2=threading.Thread(target=task2)
t1.start()
t2.start()
t1.join()
t2.join()
print("all task is complete")
```

# MULTIPROCESSING(Run only this code in vs code / co-lab)

Multiprocessing is a technique where multiple process run in parallel,each with it's own memory space to fully utilize to cpu core.

Multiprocessing means multiple process running at same time with it's own memory space and it's own python interpretor

```python
from multiprocessing import Process
import time
def task1():
    print("task1 is begin")
    time.sleep(1)
    print("task1 is end")
def task2():
    print("task2 is begin")
    time.sleep(1)
    print("task2 is end")
if __name__=="__main__":
    p1=Process(target=task1)
    p2=Process(target=task2)
    p1.start()
    p2.start()
    p1.join()
    p2.join()
    print("all task is done")
```

# Multithreading vs Multiprocessing

Multithreading is used for I/O bound tasks and shares memory but is limited by GIL(Global Interpreter Lock).

Multiprocessing is used for cpu bound tasks and achieve true parallelism uses seprate memory space.

#Decorator

A decorator in Python is a special function that allows you to modify or enhance another function’s behavior without permanently changing its code.

```python
def decorator_function(original_function):
    def wrapper_function():
        print("start")
        original_function()
        print("end")
    return wrapper_function
@decorator_function
def say_hello():
    print("hello world")
say_hello()
```

# Generator

In python generator is a special type of function that produces value one at a time using "yield" keyword instead of using "return".

It allows you to iterate through potentially large or infinite sequences without storing everything in memory, making it highly efficient.

```python
def my_generator():
    yield 1
    yield 2
    yield 3
for num in my_generator():
    print(num)
```

# ITERATOR

In Python, an iterator is an object that allows you to traverse through a sequence of elements one at a time.

  __iter__() → returns the iterator object itself

  __next__() → returns the next element in the sequence, and raises StopIteration when no more elements are left.

```python
class countdown:
    def __init__(self,start):
        self.start=start
    def __iter__(self):
        return self
    def __next__(self):
        if self.start<=0:
            raise StopIteration
        self.start-=1
        return self.start
cd=countdown(7)
for num in cd:
    print(num)
```

# Abstraction<<

abstraction in python hiding the internal implementation details and showing only essential feature to the users

```python
from abc import ABC, abstractmethod

class Shape(ABC):   # Abstract class
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):   # Concrete class
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Object creation
rect = Rectangle(10, 5)
print("Area:", rect.area())          # Output: 50
print("Perimeter:", rect.perimeter()) # Output: 30
```

# Encapsulation

Encapsulation is a fundamental principle of object-oriented programming (OOP). It means bundling data (attributes) and the methods that operate on that data into a single unit — typically a class — while restricting direct access to some of the object's components.  
This helps protect the internal state of an object and ensures that it can only be modified in controlled ways.

```python
class Employee:
    def __init__(self, name, project, salary) -> None:
        self.name = name          # Public (Accessible everywhere)
        self._project = project   # Protected (Accessible within class or subclass)
        self.__salary = salary    # Private (Accessible only inside class)

    def detail(self):
        print(f"name is : {self.name}")    
        print(f"project is : {self._project}")
        print(f"salary is : {self.__salary}")

# Object creation
Object = Employee("Rohit", "Python", 100000)

# Access through method
Object.detail()

# Public and protected can be accessed directly
print(Object.name)       # ✅ Output: Rohit
print(Object._project)   # ✅ Output: Python

# Private variable cannot be accessed directly
# print(Object_rohit.__salary)   # ❌ Error

# Correct way to access private variable (name mangling)
print(Object._Employee__salary)  # ✅ Output: 100000
```

# Polymorphism 

The word polymorphism means having many forms. In programming, polymorphism means the same function name (but different signatures) being used for different types. The key difference is the data types and number of arguments used in function.



```python
class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# Runtime polymorphism in action
animals = [Dog(), Cat(), Animal()]
for a in animals:
    print(a.speak())
```

# Inheritance

Inheritance is an OOP concept where a child class acquires the properties and methods of a parent class, enabling code reusability and establishing a hierarchical relationship between classes.

```python
# Parent class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}")

# Child class inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        # Call parent constructor
        super().__init__(brand, model)
        self.doors = doors
    
    def display_info(self):
        # Extend parent method
        super().display_info()
        print(f"Doors: {self.doors}")

# Usage
my_car = Car("Toyota", "Corolla", 4)
my_car.display_info()
```
